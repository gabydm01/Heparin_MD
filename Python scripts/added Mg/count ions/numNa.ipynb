### This script plots the number of ions adsorbed and its average value

```python
import numpy as np
import matplotlib.pyplot as plt 
import scienceplots
plt.style.use('science')

dat = np.loadtxt('numNa.dat')
count = dat[:,1] 
frame = dat[:,0]
t = frame/10
n = np.zeros(len(count))
n = n+np.mean(count)
nmean = np.mean(count)
plt.figure(figsize = (9,6))
plt.plot(t, count, c = 'darkorchid')
plt.xlabel('t (ns)', fontsize = 22)
plt.ylabel("Number of ions", fontsize = 22)
plt.plot(t, n, c = 'r', label = 'Average = {:.3f}'.format(nmean))
plt.legend(fontsize = 14)
plt.savefig('numNa_dimer.png')
print('The average shielded charge is {:.3f}'.format(nmean))
```
