### This script plots the number of ions adsorbed and its average value

```python
import numpy as np
from numpy import sqrt, pi, exp, linspace, loadtxt, power
import matplotlib.pyplot as plt 
import scienceplots
plt.style.use('science')

dat = loadtxt('numMg.dat')
count = dat[:,1] 
frame = dat[:,0]
t = frame/10
n = np.zeros(len(count))
n = n+np.mean(count)
nmean = np.mean(count)

plt.figure(figsize = (9,6))
plt.plot(t, count, c = 'darkorchid')
plt.xlabel('t (ns)', fontsize = 22)
plt.ylabel("Number of ions", fontsize = 22)
plt.plot(t, n, c = 'r', label = 'Average = {:.3f}'.format(nmean))
plt.legend(fontsize = 14)
plt.savefig('numMg.png')
print('The average shielded charge is 2*{:.3f} = {:.3f}'.format(nmean,nmean*2))
```
