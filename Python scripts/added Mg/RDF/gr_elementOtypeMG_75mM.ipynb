### This script plots the RDF $g(r)$ of the oxygens of heparin and the magnesium ions

```python
import numpy as np
from numpy import sqrt, pi, exp, linspace, loadtxt, power
import matplotlib.pyplot as plt 
import scienceplots
plt.style.use('science')

G = loadtxt('gr_elementOtypeMG_75mM.dat')
g = G[:,1] 
r = G[:,0] 
gmax = np.max(g)


plt.figure(figsize = (9,6))
plt.plot(r, g, c = 'darkorchid',  label = 'element O and not water - type MG')
# plt.title('$g(r)$ element O (not water) - type MG')
for i in range(len(g)):
    if g[i] == gmax:
        rmax = r[i]
        plt.axvline(x=rmax,ymax = .95, c = 'k', ls = ':')
plt.ylabel('$g(r)$', fontsize = 22)
plt.xlabel('$r$ (\\r{A})', fontsize = 22)
plt.axvline(x=5.05,ymax = .35, c = 'k', ls = ':')
plt.legend(fontsize = 14)

print('Maximum correlation occurs at r = {:.3f} Å'.format(rmax))
print('First coordination shell at r < 5.05 Å')
```
