### This script plots the RDF $g(r)$ of the oxygens of heparin and the sodium ions

```python
import numpy as np
import matplotlib.pyplot as plt 
import scienceplots
plt.style.use('science')

G = np.loadtxt('gr_elementOtypeSOD.dat')
g = G[:,1] 
r = G[:,0] 
gmax = np.max(g)

plt.figure(figsize = (9,6))
plt.plot(r, g, c = 'darkorchid', label = 'element O and not water - type SOD')
# plt.title('$g(r)$ element O (not water) - type SOD')
for i in range(len(g)):
    if g[i] == gmax:
        rmax = r[i]
        plt.axvline(x=rmax,ymax = .95, c = 'k', ls = ':')
plt.ylabel('$g(r)$', fontsize = 22)
plt.xlabel('$r$ (\\r{A})', fontsize = 22)
plt.axvline(x=3.2,ymax = .075, c = 'k', ls = ':')
plt.legend(fontsize = 14)
plt.savefig('g_r_Na_dimer.png')

print('Maximum correlation occurs at r = {} Å'.format(rmax))
print('First coordination shell at r < 3.2 Å')
```
